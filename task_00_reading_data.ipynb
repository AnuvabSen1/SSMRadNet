## Reading Images

```python
import cv2
import matplotlib.pyplot as plt

# 1) Point to your .mjpg file
video_path = "RECORD@2020-11-22_12.42.37/RECORD@2020-11-22_12.42.37_camera.mjpg"

# Target max height for 480p
MAX_HEIGHT = 480

# Open capture
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise IOError(f"Cannot open {video_path}")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Get original dimensions
    h, w = frame.shape[:2]
    # Compute new dimensions preserving aspect ratio
    new_h = MAX_HEIGHT
    new_w = int(w * (new_h / h))

    # Resize frame
    frame_resized = cv2.resize(
        frame,
        (new_w, new_h),
        interpolation=cv2.INTER_AREA
    )

    # Display
    cv2.imshow("480p MJPG Playback", frame_resized)
    # waitKey(30) ~ 33 FPS; press 'q' to quit
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

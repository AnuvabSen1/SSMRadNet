### Download using service script - with checkpointing

```python
import os
import json
from pydrive2.auth import GoogleAuth
from pydrive2.drive import GoogleDrive
from tqdm import tqdm

def authenticate_service_account(credentials_path='service_account_credentials.json'):
    settings = {
        "client_config_backend": "service",
        "service_config": {
            "client_json_file_path": credentials_path
        }
    }
    gauth = GoogleAuth(settings=settings)
    gauth.ServiceAuth()
    return GoogleDrive(gauth)

def scan_drive_files(drive, folder_id, folder_path=""):
    """Recursively scan and list all files under a Google Drive folder."""
    file_list = drive.ListFile({'q': f"'{folder_id}' in parents and trashed=false"}).GetList()
    entries = []

    for file in file_list:
        name = file['title']
        full_path = os.path.join(folder_path, name)
        if file['mimeType'] == 'application/vnd.google-apps.folder':
            entries += scan_drive_files(drive, file['id'], full_path)
        else:
            entries.append({
                'id': file['id'],
                'name': name,
                'path': full_path,
                'size': int(file.get('fileSize', 0)),
                'status': 'pending'
            })
    return entries

def load_checkpoint(json_path):
    if os.path.exists(json_path):
        with open(json_path, 'r') as f:
            return json.load(f)
    return []

def save_checkpoint(json_path, data):
    with open(json_path, 'w') as f:
        json.dump(data, f, indent=2)


def download_files(drive, file_list, output_dir, checkpoint_path):
    for file in tqdm(file_list, desc="Downloading files", unit="file"):
        if file['status'] == 'done':
            continue

        save_path = os.path.join(output_dir, file['path'])
        os.makedirs(os.path.dirname(save_path), exist_ok=True)

        try:
            f = drive.CreateFile({'id': file['id']})
            f.GetContentFile(save_path)
            tqdm.write(f"Downloaded: {file['path']}")
            file['status'] = 'done'
            save_checkpoint(checkpoint_path, file_list)  # update after each file
        except Exception as e:
            tqdm.write(f"Failed: {file['path']} | Error: {str(e)}")

            

            
root_folder_id = "1ur9Dgr7fjE4v0QbdBKXLtk5t8ey__3wn"
output_dir = "./RaDIAL_dataset/raw_sequences"
checkpoint_path = "download_checkpoint.json"

drive = authenticate_service_account()

# Load or scan
file_list = load_checkpoint(checkpoint_path)
if not file_list:
    print("Scanning Drive folder...")
    file_list = scan_drive_files(drive, root_folder_id)
    save_checkpoint(checkpoint_path, file_list)

# Download all pending
download_files(drive, file_list, output_dir, checkpoint_path)
```
